```python
import os
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import mne
import re
import glob
from mne.preprocessing import find_bad_channels_lof
os.chdir(r'D:\Thesis\ThesisEEG\ds003969-download')
```

```python
sub='01'
task='med2'
base_path = r"D:\Thesis\ThesisEEG\ds003969-download"
bdf_path = fr"{base_path}\sub-{sub}\eeg\sub-{sub}_task-{task}_eeg.bdf"
```

```python
paths = glob.glob(fr"{base_path}\sub-*\eeg\sub-*_task-{task}_eeg.bdf")
records = []

def detect_bad_channels_lof(bdf_path, l_freq=1, h_freq=40):
    # Extract subject ID from THIS file only
    filename = os.path.basename(bdf_path)
    match = re.search(r"sub-(\d+)", filename)
    subject = match.group(1) if match else "unknown"

    # 1. Load raw BDF
    raw = mne.io.read_raw_bdf(bdf_path, preload=True, verbose=False)
    raw.resample(250)
    raw = raw.pick_types(eeg=True, exclude='bads')

    # 2. Preprocessing
    raw.filter(l_freq, h_freq)
    raw.set_eeg_reference('average', projection=False)

    # 3. LOF bad-channel detection
    bads, scores = find_bad_channels_lof(raw, return_scores=True)
    raw.info['bads'] = bads

    # Store results
    records.append({
        "subject": subject,
        "bad_channels": ", ".join(bads) if bads else ""
    })

    return raw, bads, scores
```

```python
results = {}
for p in paths:
    raw, bads, scores = detect_bad_channels_lof(p)
    results[p] = bads
    print(p, bads)
```

```python
df = pd.DataFrame(records)
df.to_excel("bad_channels.xlsx", index=False)
```

```python
import pandas as pd
```

```python
df= pd.read_excel('ds003969-download/bad_channels.xlsx')
```

```python
def parse_list(x):
    if pd.isna(x) or x == "":
        return []
    return [ch.strip() for ch in str(x).split(",")]

df["bad_channels"] = df["bad_channels"].apply(parse_list)
df
```

```python
df["n_bad"] = df["bad_channels"].apply(len)
```

```python
total_bad_channels = df["n_bad"].sum()
total_bad_channels
```

```python
df["n_bad"]
percent_with_bad = (df["n_bad"] > 0).mean() * 100
percent_with_bad
average_bad_per_subject = df["n_bad"].mean()
average_bad_per_subject
```

```python
import mne
import matplotlib.pyplot as plt
from scipy.signal import detrend, welch 
import os
import pandas as pd 
import numpy as np
```

```python
def compute_alpha2_bandpower(raw, win=1): #hamming window =1s
    data = raw.get_data()
    sfreq = raw.info['sfreq']
    nperseg = int(sfreq * win)

    freqs, psd = welch(
        data,
        fs=sfreq,
        nperseg=nperseg,
        window='hamming',
        axis=1
    )

    # Alpha-2 band
    f_low, f_high = 10, 12
    idx = (freqs >= f_low) & (freqs <= f_high)

    band_power = np.trapz(psd[:, idx], freqs[idx], axis=1)
    total_power = np.trapz(psd, freqs, axis=1)

    rel_power = band_power / total_power
    return rel_power.mean()  # average across channels
```

```python
data_dir = "D:\\Thesis\\ThesisEEG\\datameditation\\processed"
```

```python
results = []    
for fname in os.listdir(data_dir):
    if fname.endswith("_eeg_clean_raw.fif"):

        # Extract subject and task correctly
        parts = fname.split("_")
        subject = parts[0]  # e.g., "sub-001"
        task_name = parts[1].replace("task-", "")  # e.g., "rest" or "meditation"    

        raw = mne.io.read_raw_fif(os.path.join(data_dir, fname), preload=True)
        alpha2_value = compute_alpha2_bandpower(raw)

        results.append({
            "subject": subject,
            "task": task_name,
            "alpha2_relative_bandpower": alpha2_value
        })

# Convert to DataFrame
df = pd.DataFrame(results)
```

```python
output_path = "D:\\Thesis\\ThesisEEG\\alpha2meditation_results.xlsx"
df.to_excel(output_path, index=False)

print(df)
```
